```python

def fxn(column):
    mean=column.mean()
    std=column.std()
    z=(column-mean)/std
    return column[(z>3) | (z<-3)]

print(fxn(numeric["cholestoral"]))
```

### outlier removal in nrml dis d.s 

```python

def fxn(column): 
    mean=column.mean()
    std=column.std()
    z=(column-mean)/std
    return column[(z>3) | (z<-3)]
outlier=fxn(numeric["cholestoral"]) 
f1=numeric['cholestoral'].isin(outlier) # outliers stored in f1
numeric=numeric[~f1] 
numeric=numeric.reset_index(drop=True)
print(numeric.shape)
```

```python

category=category[~f1]
category=category.reset_index(drop=True)
print(category.shape)
```

```python
def fxn2(column): 
    mean=column.mean()
    std=column.std()
    z=(column-mean)/std
    return column[(z>3) | (z<-3)]
outlier=fxn2(numeric["oldpeak"]) 
f2=numeric['oldpeak'].isin(outlier) # outliers stored in f
numeric=numeric[~f2] 
numeric=numeric.reset_index(drop=True)
print(numeric.shape)
```

```python
category=category[~f2]
category=category.reset_index(drop=True)
print(category.shape)
```

```python
final_df=pd.concat((numeric,category),axis=1)
final_df


```

#### split data

```python


x=final_df.drop(columns=["target"])
y=final_df["target"]
x
```

#### scaling

```python


from sklearn.preprocessing import StandardScaler
sk=StandardScaler()
x=sk.fit_transform(x)
x
```

#### splitting of data 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(y_test)
```

```python

from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(x_train,y_train)
y_pred=log.predict(x_test)
print(y_test) # pred
```

### R.f.classifier

```python

from sklearn.ensemble import RandomForestClassifier
d=RandomForestClassifier()
d.fit(x_train,y_train)
y_pred=d.predict(x_test)
print(y_test,y_pred)
print(y_test)

```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
print(confusion_matrix(y_test,y_pred))
```
